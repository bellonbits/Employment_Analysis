```python
import pandas as pd
import matplotlib as pl
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('pop.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
m_vs_f = df.groupby('Years').agg({'Unemployment_youth_male':'sum', 'Unemployment_youth_female':'sum'}).reset_index()
```

```python
m_vs_f
```

```python
# Plot the graph
m_vs_f.plot(x='Years', y=['Unemployment_youth_male', 'Unemployment_youth_female'], kind='line', marker='o')

# Add labels and title
plt.xlabel('Years')
plt.ylabel('Unemployment Count')
plt.title('Youth Unemployment (Male vs Female) Over Years')

# Show the plot
plt.show()
```

```python
# Plot the graph
df.plot(x='Years', y='Population', kind='line', marker='o')

# Add labels and title
plt.xlabel('Years')
plt.ylabel('Population Count')
plt.title('Population Over Years')

# Show the plot
plt.show()
```

```python
emp = df.groupby('Years').agg({'perc_male_empl':'mean', 'perc_female_empl':'mean'}).reset_index()
```

```python
emp
```

```python
emp.dropna()
```

```python
emp
```

```python
# Plot the graph
emp.plot(x='Years', y=['perc_male_empl', 'perc_female_empl'], kind='bar')

# Add labels and title
plt.xlabel('Years')
plt.ylabel('% of employment')
plt.title('Youth employment percentage (Male vs Female) Over Years')

# Show the plot
plt.show()
```
